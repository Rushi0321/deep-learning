```python
import matplotlib.pyplot as plt
def plot_sample(x,y,index):
    plt.figure(figsize=(10,6))
    plt.matshow(x[index])
    plt.xlabel(classes[y[index]])
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
#in cnn do following stepos...
```

```python
"""1)check for null values 
2)check for y_train,y_test shape and if shape is 2D then reshape it to 1d by y=y.reshape(-1)
3)also the scale thae value in x_train anjd x_test by x_train/255 and x_test/255"""
```

```python
#now building the model...
```

```python
from tensorflow.keras import layers
model_cnn=keras.Sequential([
    layers.Conv2D(filters=32,kernel_size=(3,3),activation="relu",input_shape=(32,32,3)),
    layers.MaxPool2D((2,2)), #maxpooling function absrtrct max information from op image for less layers...maxpool size is 2*2 
    layers.Conv2D(filters=64,kernel_size=(3,3),activation="relu"),
    layers.MaxPool2D((2,2)),
    layers.Flatten(),
    layers.Dense(10,activation='sigmoid')
])
```

```python
model_cnn.compile(optimizer='adam',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])
model_cnn.fit(x_train,y_train,epochs=10)
```

```python
model_cnn.evaluate(x_test,y_test)
```
